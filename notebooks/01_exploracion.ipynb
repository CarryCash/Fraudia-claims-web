# 01 – Exploración de datos sintéticos
En este notebook inspeccionamos el conjunto de datos generado en el paso anterior.
Se muestra un *quick look* y algunas visualizaciones básicas.

```python
import pandas as pd, numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Cargar el CSV generado
df = pd.read_csv('data/raw/siniestros.csv', parse_dates=['fecha_ocurrencia','fecha_reporte'])
df.head()
```

```python
# Información general
df.info()
```

```python
# Estadísticas descriptivas de variables numéricas
df.describe().T
```

```python
# Distribución de montos reclamados
plt.figure(figsize=(8,4))
sns.histplot(df['monto_reclamado'], bins=50, kde=True, color='steelblue')
plt.title('Distribución del monto reclamado')
plt.xlabel('Monto ($)')
plt.show()
```

```python
# Análisis de retraso entre ocurrencia y reporte
df['dias_entre_ocurrencia_reporte'] = (df['fecha_reporte'] - df['fecha_ocurrencia']).dt.days
sns.boxplot(x='dias_entre_ocurrencia_reporte', data=df, color='orange')
plt.title('Días entre ocurrencia y reporte')
plt.show()
```
